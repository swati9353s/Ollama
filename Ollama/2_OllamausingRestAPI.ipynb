# GENERATE USING OLLAMA LIBRARY

```python
#Generating using the ollama library
import ollama

response= ollama.generate(
    model="llama3.2:1b",
    prompt="Explain black holes simply"
)

print(response.response)
```

# GENERATE USING REST API

```python
import requests   # Used to send HTTP requests (GET, POST, etc.) to APIs
import json       # Used to work with JSON data (encode/decode JSON format)

# URL of the local API endpoint (Ollama server running locally)
url = "http://localhost:11434/api/generate"

# Payload = data we send to the API in JSON format
payload = {
    "model": "llama3.2:1b",          # Name of the LLM model to use
    "prompt": "Explain black holes simply"   # Input text/question sent to the model
}

# requests.post()
# → Sends an HTTP POST request to the given URL
# → json=payload automatically converts the Python dictionary
#   into JSON and sends it in the request body
response = requests.post(url, json=payload)

# The server processes the request and returns a response object
# response contains:
#   - status code (success/failure)
#   - headers
#   - generated output from the model (inside response.text or response.json())
```

```python
# Variable to store the final generated text from the model
output = ""

# iter_lines()
# → Reads the API response line-by-line (streaming response)
# → Useful when the model sends output gradually instead of all at once
for i in response.iter_lines():

    # Skip empty lines
    if i:
        # Decode bytes → string and convert JSON string → Python dictionary
        data = json.loads(i.decode("utf-8"))

        # Each streamed chunk contains part of the generated text
        # Append the partial response to the final output
        if "response" in data:
            output += data["response"]

        # "done" becomes True when generation is complete
        # Break the loop once full response is received
        if data.get("done"):
            break

# Print the complete generated output
print(output)
```
